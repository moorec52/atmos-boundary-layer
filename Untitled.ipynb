```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cues = pd.read_pickle('data/CUES_L1_Temp_Rad_Wind_Mar-Apr2017.pkl')
```

```python
cues.columns
```

```python
cues['sonic wind direction; Sonic anemometer; (degrees)'].plot()
cues['Sonic wind direction  standard deviation; Sonic anemometer; (degrees)'].plot()
cues['wind quality; ws600 wind quality; (percent)'].plot()
cues['sonic wind speed; Sonic anemometer; (meters/second)'].plot()
```

```python
startdate = pd.Timestamp('2017-03-10')
enddate = pd.Timestamp('2017-03-11')

fig, ax = plt.subplots(2,1,figsize=(10,7))
# Snow surface temp
ax[0].set_title('Snow Surface Temperature\n{}-{}'.format(startdate,enddate))
cues['snow surface temperature; Apogee SI-111 thermal radiometer; (Celsius)'].plot(ax=ax[0],
                                                                                  c='k')
ax[0].set_ylim((-10,1))

# Air temp
ax[1].set_title('Air Temperature\n{}-{}'.format(startdate,enddate))
cues['Platform temperature; platform temperature/relative humidity; (degrees Celsius)'].plot(ax=ax[1],
                                                                                            c='k')
ax[1].set_ylim((-1,10))


# finish up plot labels, axes, etc
for _ax in ax:
    _ax.set_xlim((startdate,enddate))
    #_ax.set_ylim((-15,15))
    _ax.set_ylabel('Temperature [$C\degree$]')
    _ax.set_xlabel('Date, Time')
    _ax.grid(axis='x',color='lightgrey', linestyle=':', linewidth=1)
    _ax.axhline(0,color='k', linestyle='--', linewidth=1, zorder=-99)

fig.autofmt_xdate(rotation=0, ha='center')
plt.tight_layout()
```

```python
winds = pd.read_csv(r'\\j-lundquist-3.ce.washington.edu\storage\CUES\CUESWinds201703through201704\CUESWinds201703through201704.csv')
winds['datetime']  = pd.to_datetime(winds['TIMESTAMP'])
winds = winds.set_index('datetime')
```

```python
startdate = pd.Timestamp('2017-03-11')
enddate = pd.Timestamp('2017-03-12')

plt.figure(figsize=(5,5))

winds.Temp.plot()
cues['Platform temperature; platform temperature/relative humidity; (degrees Celsius)'].plot()
cues['snow surface temperature; Apogee SI-111 thermal radiometer; (Celsius)'].plot()

plt.ylim((-15,15))
plt.xlim((startdate,enddate))
```

```python
winds.Sonic3D_U.plot()
winds.Sonic3D_V.plot()
winds.Sonic3D_W.plot()
```

```python
startdate = pd.Timestamp('2017-03-09')
enddate = pd.Timestamp('2017-03-12')

fig, ax = plt.subplots(5,1,figsize=(15,15))
# Snow surface temp
ax[0].set_title('Snow Surface Temperature\n{}-{}'.format(startdate,enddate))
cues['snow surface temperature; Apogee SI-111 thermal radiometer; (Celsius)'].plot(ax=ax[0],
                                                                                  c='k')
ax[0].set_ylim((-10,1))

# Air temp
ax[1].set_title('Air Temperature\n{}-{}'.format(startdate,enddate))
cues['Platform temperature; platform temperature/relative humidity; (degrees Celsius)'].plot(ax=ax[1],
                                                                                            c='k')
ax[1].set_ylim((-1,10))

# finish up plot labels, axes, etc
for _ax in ax[0:2]:
    _ax.set_xlim((startdate,enddate))
    #_ax.set_ylim((-15,15))
    _ax.set_ylabel('Temperature [$C\degree$]')
    _ax.set_xlabel('Date, Time')
    _ax.grid(axis='x',color='lightgrey', linestyle=':', linewidth=1)
    _ax.axhline(0,color='k', linestyle='--', linewidth=1, zorder=-99)

# 3D winds
winds.Sonic3D_U.plot(ax=ax[2], label='wind velocity')
ax[2].fill_between(winds.index, winds.Sonic3D_U + winds.Sonic3D_U_SD, winds.Sonic3D_U - winds.Sonic3D_U_SD,
                  alpha=0.5, label='+/- 1 sd')
ax[2].set_ylabel('U [m/s]')
ax[2].set_ylim((-5,5))

winds.Sonic3D_V.plot(ax=ax[3], label='wind velocity')
ax[3].fill_between(winds.index, winds.Sonic3D_V + winds.Sonic3D_V_SD, winds.Sonic3D_V - winds.Sonic3D_V_SD,
                  alpha=0.5, label='+/- 1 sd')
ax[3].set_ylabel('V [m/s]')
ax[3].set_ylim((-5,5))

winds.Sonic3D_W.plot(ax=ax[4], label='wind velocity')
ax[4].fill_between(winds.index, winds.Sonic3D_W + winds.Sonic3D_W_SD, winds.Sonic3D_W - winds.Sonic3D_W_SD,
                  alpha=0.5, label='+/- 1 sd')
ax[4].set_ylabel('W [m/s]')
ax[4].set_ylim((-5,5))

# finish up plot labels, axes, etc
for _ax in ax[2:]:
    _ax.set_xlim((startdate,enddate))
    _ax.set_ylim((-5,5))
    _ax.set_xlabel('Date, Time')
    _ax.grid(axis='both',color='lightgrey', linestyle=':', linewidth=1)
    _ax.axhline(0,color='k', linestyle='--', linewidth=1, zorder=-99)
    _ax.legend(loc='upper left')

# more plot formatting
fig.autofmt_xdate(rotation=0, ha='center')
plt.tight_layout()
```
